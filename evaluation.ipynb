```python
#untuk membuat verses menjadi list
file_arab['verses_list'] = file_arab.apply(lambda row: row['verses'].split(', '), axis=1)
file_arab
```

```python
#untuk membuat verses menjadi list
file_arab_goldstandard['verses_list'] = file_arab_goldstandard.apply(lambda row: row['verses'].split(', '), axis=1)
file_arab_goldstandard
```

```python
df1 = pd.DataFrame(file_arab.verses_list.tolist(), index=file_arab.lcs) \
        .stack() \
        .reset_index(level=1, drop=True) \
        .reset_index(name='verses_list')[['verses_list','lcs']]
print (df1)
```

```python
df2 = pd.DataFrame(file_arab_goldstandard.verses_list.tolist(), index=file_arab_goldstandard.lcs) \
        .stack() \
        .reset_index(level=1, drop=True) \
        .reset_index(name='verses_list')[['verses_list','lcs']]
print (df2)
```

```python
#fungsi untuk mengubah dataframe menjadi set
program_set = set(df1.apply(lambda row: row['lcs'] + ' ' + row['verses_list'], axis=1))
program_set
```

```python
#fungsi untuk mengubah dataframe menjadi set
gold_set = set(df2.apply(lambda row: row['lcs'] + ' ' + row['verses_list'], axis=1))
gold_set
```

```python
tp = len(program_set.intersection(gold_set))
tp
```

```python
set_fp = program_set.difference(gold_set)
fp = len(set_fp)
fp
```

```python
set_fn = gold_set.difference(program_set)
fn = len(set_fn)
fn
```

```python
#tingkat kebenaran
precision = tp/(tp+fp) * 100
print(precision,"%")
```

```python
#tingka kelengkapan
recall = tp/(tp+fn) * 100
print(recall,"%")
```

```python
#akurasi sistem yang dibuat
f1_score = (2*(recall * precision) / (recall + precision))
f1_score
```
