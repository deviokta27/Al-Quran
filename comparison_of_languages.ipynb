```python
import pandas as pd
```

```python
file_indo = pd.read_excel("data_surat1-5_indo.xlsx")
file_arab = pd.read_excel("data_surat1-5_arab_lem.xlsx")
```

```python
file_indo
```

```python
file_arab
```

```python
ayat_indo= file_indo['verses']
```

```python
ayat_indo
```

```python
ayat_arab= file_arab['verses']
```

```python
ayat_arab
```

```python
persamaan= pd.Series(list(set(ayat_indo) & set(ayat_arab)))
# inplace fungsinya untuk memasukan nilai ke variable awalnya
persamaan.sort_values(inplace=True)
persamaan.reset_index(drop=True, inplace=True)
persamaan
```

```python
persamaan
```

```python
persamaan_indo = file_indo[file_indo['verses'].isin(persamaan)]
persamaan_indo.sort_values(by='verses', inplace=True)
persamaan_indo.reset_index(inplace=True)
persamaan_indo
```

```python
persamaan_arab = file_arab[file_arab['verses'].isin(persamaan)]
persamaan_arab.sort_values(by='verses', inplace=True)
persamaan_arab.reset_index(inplace=True)
persamaan_arab
```

```python
persamaan_indo['verses']
```

```python
df_persamaan = pd.DataFrame({
    'verses': persamaan,
    'verse_count': persamaan_indo['verse_count'],
    'lcs_indo': persamaan_indo['lcs'], 
    'jumlah_kata_indo': persamaan_indo['jumlah_kata'],
    'lcs_arab': persamaan_arab['lcs'], 
    'jumlah_kata_arab': persamaan_arab['jumlah_kata']})
df_persamaan
```

```python
from pandas import ExcelWriter

writer = ExcelWriter('data_perbandingan_arab_indo_surat1-5.xlsx')
df_persamaan.to_excel(writer,'perbandingan')
writer.save()
```
